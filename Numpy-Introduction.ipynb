# NumPy gyakorló feladat – 3×3 mátrix

## Feladat

1. **Hozd létre az alábbi mátrixot** `A` néven a következő értékekkel:

|   |   |   |
|---|---|---|
| 4  | 11 | 7  |
| 9  | 5  | 12 |
| 3  | 14 | 8  |

> Tipp: használhatod a `np.array()` függvényt.

2. Szorozd meg a **második sort 2-vel**.  
3. Vonj ki **4-et a harmadik oszlop minden eleméből**.  
4. Állítsd a **harmadik sor, harmadik elemet 100-ra**.  
5. Adj **+2-t minden olyan elemhez, ami nagyobb 10-nél**.  
6. Számold ki **minden sor összegét**.

---

## Megoldás utáni mátrix

|   |   |   |
|---|---|---|
| 4   | 13  | 3   |
| 20  | 10  | 22  |
| 3 | 16  | 102 |

### Sorösszegek:
20, 52 121


```python
import numpy as np

a = np.array([[4, 11, 7], [9, 5, 12], [3, 14, 8]])
a[1, :] *= 2
a[:, 2] -= 4
a[2, 2] = 100
a[a > 10] += 2

print(f"A megoldás: \n {a} \n")

print(f"Sorösszegek: \n {a.sum(axis=1)}")
```
